```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
observations_df = pd.read_csv('observations.csv')
```

```python
species_info_df = pd.read_csv('species_info.csv')
species_info_df.head()
```

```python
df = pd.merge(observations_df, species_info_df, on='scientific_name')
df.head()
```

```python
conservation_counts = df['conservation_status'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=conservation_counts.index, y=conservation_counts.values, palette='viridis')
plt.title('Distribution of Conservation Status for Animals')
plt.xlabel('Conservative Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('conservation_distribution.png', dpi=300)
plt.show()
```

```python
max_observations = df.loc[df.groupby('park_name')['observations'].idxmax()]

plt.figure(figsize=(20,12))
sns.barplot(x='observations', y='park_name', hue='common_names', data=max_observations, palette='viridis')
plt.title('Species with the Most Observations at Each Park')
plt.xlabel('Number of Observations')
plt.ylabel('Park Name')
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('species_park.png', dpi=300)

plt.show()
```

```python
# Yellowstone contains most endangered species. 
```

```python
endangered_speces = df[df['conservation_status'] == 'Endangered']

plt.figure(figsize=(14,8))
sns.countplot(y='common_names', data=endangered_speces,palette='Reds', order=endangered_speces['common_names'].value_counts().index)
plt.title('Endangered Animals')
plt.xlabel('Count')
plt.ylabel('Animal Species')
plt.tight_layout()
plt.savefig('most_endangered.png',dpi=300)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='park_name', hue='conservation_status', data=df, palette='viridis')
plt.title('Proportion of Conservation Status by Park')
plt.xlabel('Park Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(y='common_names', data=df, palette='coolwarm', order=df['common_names'].value_counts().head(10).index)
plt.title('Top 10 Most Observed Species')
plt.xlabel('Count')
plt.ylabel('Species')
plt.show()
```

```python
# Group by category and conservation_status
category_conservation = df.groupby(['category', 'conservation_status']).size().reset_index(name='count')

# Calculate the total species in each category
total_category = df['category'].value_counts().reset_index()
total_category.columns = ['category', 'total']

# Merge the counts with the total to get the proportion
category_conservation = pd.merge(category_conservation, total_category, on='category')
category_conservation['proportion'] = category_conservation['count'] / category_conservation['total']

# Plot the data
plt.figure(figsize=(14, 8))
sns.barplot(x='proportion', y='category', hue='conservation_status', data=category_conservation, palette='viridis')
plt.title('Proportion of Species by Conservation Status and Category')
plt.xlabel('Proportion')
plt.ylabel('Category')
plt.legend(title='Conservation Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('endangered_by_species.png', dpi=300)
plt.show()
```
